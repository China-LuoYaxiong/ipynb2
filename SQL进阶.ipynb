# 日期维度表
- 经常需要计算当周 当月 当季 当年这些指标，就需要知道日期对应的当周周一的日期等，比较麻烦，可以首先建立一个日期维度表，随时可以用

````python
start_date = '2024-05-06'
lenth = 2
df = mc.select_data(f'''

select 
    cast(date as string)                          as date
    ,cast(replace(date, '-', '') as string)       as date_ds
    ,year(date)                                   as year
    ,quarter(date)                                as quarter
    ,month(date)                                  as month
    ,substr(date, 1, 7)                           as ym
    ,weekofyear(date)                             as weekofyear
    ,day(date)                                    as day
    
    -- 星期几,dayofweek返回1-7,1代表星期日，7代表星期6,所以要减1
    ,case 
        when dayofweek(date)-1==0 then 7
        else dayofweek(date)-1
    end                                           as dayofweek  --(1~7)
    
    -- 当周周一对应的日期
    ,cast(    date_sub(date,
                            case 
                                when dayofweek(date)-1==0 then 7-1
                                else dayofweek(date)-1-1
                            end 
                      )  
              as
              string)                             as firstday_of_week                          
    
    -- 当月1号对应的日期
    ,concat(substr(date, 1, 7),'-01')             as firstday_of_month   
    
    
    
    -- 当季1号对应的日期
    ,CASE  
        WHEN MONTH(date) BETWEEN 1 AND 3 THEN CONCAT_WS('-', YEAR(date), '01', '01')  
        WHEN MONTH(date) BETWEEN 4 AND 6 THEN CONCAT_WS('-', YEAR(date), '04', '01')  
        WHEN MONTH(date) BETWEEN 7 AND 9 THEN CONCAT_WS('-', YEAR(date), '07', '01')  
        ELSE CONCAT_WS('-', YEAR(date), '10', '01')  
    END                                           as firstday_of_quarter 
    
    -- -----------------------------------------------------------------------------------------
    
    -- 当年第一天对应的日期
    ,concat(substr(date, 1, 4),'-01-01')          as firstday_of_year
     
     
    -- 近3天（含当日）中的第一天的日期
    ,cast(date_sub(date, 2) as string)            as firstday_of_3d
    
    -- 近7天（含当日）中的第一天的日期
    ,cast(date_sub(date, 6) as string)            as firstday_of_7d
    
    -- 近14天（含当日）中的第一天的日期
    ,cast(date_sub(date, 13) as string)           as firstday_of_14d
    
    -- 近30天（含当日）中的第一天的日期
    ,cast(date_sub(date, 29) as string)           as firstday_of_30d
    
    -- 近60天（含当日）中的第一天的日期
    ,cast(date_sub(date, 59) as string)           as firstday_of_60d
    
    -- 近90天（含当日）中的第一天的日期
    ,cast(date_sub(date, 89) as string)           as firstday_of_90d
    
    -- --------------------------------------------------------------------------------------------------
    
    -- 周环比,上周同一天
    ,cast(date_sub(date, 7) as string)            as day_wvw
    
    -- 月环比,上月同一天
    ,cast(add_months(date, -1) as string)          as day_mvm
    
    -- 年同比,去年同一天
    ,cast(add_months(date, -12) as string)          as day_yvy
        
from 
(
    select date_add(start_date,pos) AS date
    from (
          SELECT '{start_date}' start_date
         ) t 
    lateral view posexplode(split(repeat(", ", {lenth}), ",")) tf AS pos,val
) a1


''')
````

- 两层漏斗：比如，在start_date和end_date之间，做了事假1的用户，在窗口期3天内，做了事件2的用户

### 两层漏斗
- 比如计算 2017-11-28到2017-11-29之间，做过pv的用户，在2日内也做过buy的用户漏斗

````python
start_date = '2017-11-28 00:00:00'
end_date   = '2017-11-29 23:59:59'
event1 = 'pv'

funnel_window_period = 2 # 漏斗窗口期,按天
funnel_window_period_ss = 2*24*60*60
event2 = 'cart'


# last_date表示最后漏斗完成的时间一定在这个时间之前，因为必须在窗口期内
last_date = str(pd.to_datetime(end_date)+pd.Timedelta(days=2)) 
````

````python
last_date
````

````python
sql1 = f'''

-- 在{start_date}到{end_date}之间，做过pv的用户
with t1 as 
(
    select
        uid                                       as uid
        ,product_id                               as product_id 
        ,product_category_id                      as product_category_id 
        ,behavior_type                            as behavior_type 
        ,cast(from_unixtime(event_time) as string) as {event1}_time
    from taobao_user_behavior_t1
    where 1=1
    and cast(from_unixtime(event_time) as string) between '{start_date}' and '{end_date}'
    and behavior_type='{event1}' 
),

-- 在{last_date}之前做过 buy 的用户，且做buy的时间要大于{start_date}
-- 且用户是t1中的用户
t2 as 
(
    select
        uid                                       as uid
        ,product_id                               as product_id 
        ,product_category_id                      as product_category_id 
        ,behavior_type                            as behavior_type 
        ,cast(from_unixtime(event_time) as string) as {event2}_time
    from taobao_user_behavior_t1
    where 1=1
    and behavior_type='{event2}'
    and cast(from_unixtime(event_time) as string) between '{start_date}' and '{last_date}'
    and uid in (select distinct uid from t1)
),

-- 按用户合并
t3 as 
(
    select 
         t1.uid           as uid
        ,t1.{event1}_time as {event1}_time
        ,t2.{event2}_time as {event2}_time
        
        ,datediff( to_date(t2.{event2}_time, 'yyyy-mm-dd hh:mi:ss'),
                   to_date(t1.{event1}_time, 'yyyy-mm-dd hh:mi:ss'),
                   'ss'                                                    -- 相差多少秒
         )                as 相差秒                                        -- （event2-event1）
    from t1 
    left join t2 on t1.uid=t2.uid  and t1.{event1}_time <= t2.{event2}_time
)


'''
````

````python
mc.select_data(sql1+'\n'+f'''
select * from t3 
where 1=1
and {event1}_time <= {event2}_time
limit 10
''')
````

````python
mc.select_data(sql1+'\n'+f'''

select 
    count(distinct uid) 第一层
    -- 差值要在0到2天（2*24*60*60=172800），则就是在窗口期内。
    ,count(distinct case when 相差秒 between 0 and {funnel_window_period_ss} then uid else null end ) as 第二层
from t3

-- 这里后面可以接 group by 用户属性、标签、画像等等，看不同用户群体的的漏斗转化对比
''')
````

# 留存分析  
1. 留存分析通常是指新用户在首次体验到产品核心功能的后，在后续的日期中，继续使用产品的情况。
2. 激活：通常是指让用户体验到核心功能，产生ah时刻，这个才叫激活。
3. 留存分析通常是涉及两个事件，第一个事件通常是非常重要的节点事件，且最好是用户在该事件中只会产生一次关键行为，比如注册、激活这种，第二个事件主要是为了评价用户持续使用产品，可以关注活跃等。
4. 留存的反面就是流失

## 内建窗口函数+over()

### 组内排名（3）

- 主要是进行组内的排名问题
- rank:1,2,3,3,5..
- dense_rank:1,2,3,3,4..
- row_number:1,2,3,4,5..

````python
mc.select_data('''
select 
    deptno 
    ,ename
    ,sal 
    ,rank() over (partition by deptno order by sal desc)       as rank_col
    ,dense_rank() over (partition by deptno order by sal desc) as dense_rank_col
    ,row_number() over (partition by deptno order by sal desc) as row_number_col
from window_t
''')
````

### 窗口两端值（2）
- 主要解决取组内的第一个值和最后一个值
- first_value/last_value
- 针对这个函数，窗口可以认为是组内第一行到当前行

````python
mc.select_data('''

select deptno
       ,ename 
       ,sal
       
       
       ,first_value(sal) over (partition by deptno order by sal desc) as 最内最大值 -- 降序, 开始行到当前行,取第一个,每一行取的都是窗口内最大值
       ,first_value(sal) over (partition by deptno order by sal asc)  as 组内最小值 -- 升序, 开始行到当前行,取第一个,每一行取的都是窗口内最小值
       
       -- 下面这种，因为每行的窗口是起始行-当前行，无论升序降序，当前行永远是最后一行，所以永远取自己
       ,last_value(sal) over  (partition by deptno order by sal desc) as a1
       ,last_value(sal) over  (partition by deptno order by sal asc)  as a2
       
       
from window_t;

''')
````

### 组内偏移（2）
- 主要解决，针对当前行， 取当前行前面和后面的第N条数据
- lag(expr，offset)：  返回当前行往前（朝分区头部方向）第offset行数据对应的表达式expr的值
- lead(expr，offset)： 返回当前行往后（朝分区尾部方向）第offset行数据对应的表达式expr的值

````python
mc.select_data('''

select * 
       -- lag(字段,offset),   offset要大于等于1
       ,lag(sal,1) over(partition by deptno order by sal asc) sal_前_1  -- 前面第一个值
       ,lag(sal,3) over(partition by deptno order by sal asc) sal_前_3  -- 前面第三个值
       
       ,lead(sal,1) over(partition by deptno order by sal asc) sal_后_1  -- 后面第一个值
       ,lead(sal,3) over(partition by deptno order by sal asc) sal_后_3  -- 后面第三个值
       
       
from window_t

''')
````

### 组内第N行（1）
- nth_value(字段，N)：返回表达式expr在窗口的第N条数据进行运算的结果。N大于等于1，等于1时，等价于first_value。
- 这里加了order by之后，当前行的窗口是开始行到当前行

````python
mc.select_data('''

select *
       -- 每一行的窗口是开始行，到当前行，N=1,等价于first就是指组内第一行，返回组内第一行的ename，即JACCKA
       ,nth_value(ename,1) over(partition by deptno order by sal desc) a1
       
       -- 每一行的窗口是开始行，到当前行，N=2,返回组内第2行的ename，即KING
       ,nth_value(ename,2) over(partition by deptno order by sal desc) a2
       
       -- 每一行的窗口是开始行，
       到当前行index是0 1时, 组内分别只有1行 2行，无法返回组内第三行，所以返回None
       ,nth_value(ename,3) over(partition by deptno order by sal desc) a3
       
from window_t

''')
````

### 组内分布（2）  
- cume_dit( ) :求累计分布，大小关系由orderby_clause判定。
- percent_rank()：具体的返回值等于“(rank - 1) / (partition_row_count - 1)”，  
其中：rank为该行数据的RANK窗口函数的返回结果，partition_row_count为该行数据所属分区的数据行数。当分区中只有一行数据时，输出结果为0.0。

````python
mc.select_data('''

select *,
       -- 如果升序，则统计小于等于当前值的行数/总行数
        cume_dist() over(partition by deptno order by sal asc)     as a1
        
       -- 如果降序，则统计大于等于当前值的行数/总行数
       ,cume_dist() over(partition by deptno order by sal desc)    as a2
       
       
       , rank() over(partition by deptno order by sal desc)  -1     as rk_1 
       , count(1) over(partition by deptno ) -1                     as cnt_1 
       ,(
         rank() over(partition by deptno order by sal desc)  -1
        ) 
        
        / 
        
        (
         count(1) over(partition by deptno ) -1 
        )                                                           as percent_rank_v1
       
       ,percent_rank() over(partition by deptno order by sal desc)  as percent_rank_v2
       
from window_t

''')
````

### 组内分箱（1）
- ntile(N):用于将分区中的数据按照顺序切分成N等份，并返回数据所在等份的编号。如果分区中的数据不能被均匀地切分成N等份时，最前面的等份（编号较小的）会优先多分配1条数据。
参数说明：N必填。切片数量。BIGINT类型。

````python
mc.select_data('''

select *
       ,ntile(4) over(order by sal asc)                     AS 全局分箱
       ,ntile(3) over(partition by deptno order by sal asc) AS 组内分箱

from window_t

''')
````

## 内建聚合函数+over()
- count/max/min/median/sum

````python
# 其他聚合函数是通用的
df = mc.select_data('''

select 
    deptno,
    ename,
    empno,
    sal
    
    -- 对当前行来说，窗口都是全部数据，所以是一样的。
    ,sum(sal) over()                                            AS 全局求和
    
    -- 对当前行来说，窗口开始行到当前行，所以是累计求和。
    ,sum(sal) over(order by sal asc)                            AS 全局求累计和
    
    -- 对当前行来说，窗口是每个分区的数据，同一个分区内数据是一样的。
    ,sum(sal) over(partition by deptno)                         AS 组内求和
    
    ,sum(sal) over(partition by deptno order by sal asc)        AS 组内求累计和
    
    
    -- 累计乘积,把连乘变成连加 
    ,round(
            pow(
                    10, 
                    sum( log10(sal) ) over(partition by deptno order by sal asc)
            )
    )                                                          AS 组内累计乘积
    
    
    -- 指定------------------------------------------------------------- frame_clause-------------------------------------------------------------
    -- 未显示设置frame_clause时，MaxCompute会使用默认的frame_clause来决定窗口所包含数据的边界。默认的frame_clause为：
    
    -- 1. 当开启Hive兼容模式（set odps.sql.hive.compatible=true;）时，默认的frame_clause如下，与大部分SQL系统相同。
    -- RANGE between UNBOUNDED PRECEDING and CURRENT ROW EXCLUDE NO OTHERS
    
    --2. 当关闭Hive兼容模式（set odps.sql.hive.compatible=false;），同时窗口定义中指定了order by语句
    --   且窗口函数为AVG、COUNT、MAX、MIN、STDDEV、STEDEV_POP、STDDEV_SAMP或SUM时，会使用ROWS类型的默认frame_clause。
    -- ROWS between UNBOUNDED PRECEDING and CURRENT ROW EXCLUDE NO OTHERS
    
    -- so ：默认情况下，聚合函数+over(order by xx),当前行的窗口默认就是 第一行到当前行
    
    --详细见：https://help.aliyun.com/zh/maxcompute/user-guide/window-functions-1?spm=a2c4g.11186623.0.0.78f5e14eauTA3t
    
    
    -- 常用窗口定义
    -- 1. 第一行 到 当前行(默认就是这个)
    ,sum(sal) over(partition by deptno order by sal ROWS between UNBOUNDED PRECEDING and CURRENT ROW)        求和第一行到当前行
    
    -- 2. 第一行 到 最后一行（等价于没有order by）
    ,sum(sal) over(partition by deptno order by sal ROWS between UNBOUNDED PRECEDING and UNBOUNDED FOLLOWING) 求和第一行到最后一行
    
    -- 3. 前面 第2行 到 当前行（移动平均写法）
    ,avg(sal) over(partition by deptno order by sal ROWS between 2 PRECEDING and CURRENT ROW) 近3条移动平均
    
    -- 4. 前面第1行 到 后面第一行
    ,sum(sal) over(partition by deptno order by sal ROWS between 1 PRECEDING and 1 FOLLOWING)  求和前面第1行到后面第一行
    
    
    
    -- ---------------------------------------------------加过滤条件，满足条件的才聚合-----------------------------------------
    -- 写法：聚合函数(字段) filter(where 条件1 [and/or 条件2等等]) over(partition by xx order by xxx ROWS between xxx and xxx)
    -- 当然最好还是在select .... from .... where ....中就过滤 
    
    -- 比如求，每个部门里面，名字含有C的人的工资 的累计和
    ,sum(sal) filter(where ename rlike 'C') over(partition by deptno order by sal asc)  col1
    

    
from window_t

''')
df
````

## grouping sets
- 指定要聚合的维度组合

````python
mc.select_data('''
select 
     coalesce(os, '总计')            as os
    ,coalesce(device, '总计')        as device
    ,coalesce(city, '总计')          as city
    ,count(1)                        as cnt

from requests
group by grouping sets((os, device), (city), ());
''')
````

## cube
- 特殊的grouping sets，将指定列的所有可能组合作为grouping sets，也可以与grouping sets组合使用。

```sql
group by cube (a, b, c)  
--等效于以下语句。  
grouping sets ((a,b,c),(a,b),(a,c),(b,c),(a),(b),(c),())

group by cube ( (a, b), (c, d) ) 
--等效于以下语句。 
grouping sets (
    ( a, b, c, d ),
    ( a, b       ),
    (       c, d ),
    (            )
)
group by a, cube (b, c), grouping sets ((d), (e)) 
--等效于以下语句。 
group by grouping sets (
    (a, b, c, d), (a, b, c, e),
    (a, b, d),    (a, b, e),
    (a, c, d),    (a, c, e),
    (a, d),       (a, e)
)
```

````python
mc.select_data('''
select 
     coalesce(os, '总计')            as os
    ,coalesce(device, '总计')        as device
    ,coalesce(city, '总计')          as city
    ,count(1)                        as cnt
from requests 
-- 所有可能组合的维度
group by cube (os, device, city);
''')
````

## rollup  
- 特殊的grouping sets，以按层级聚合的方式产生grouping sets，也可以与grouping sets组合使用。

```sql
group by rollup (a, b, c)
--等效价于以下语句。  
grouping sets ((a,b,c),(a,b),(a), ())

group by rollup ( a, (b, c), d ) 
--等效于以下语句。
grouping sets (
    ( a, b, c, d ),
    ( a, b, c    ),
    ( a          ),
    (            )
)
group by grouping sets((b), (c), rollup(a,b,c)) 
--等效于以下语句。 
group by grouping sets (
    (b), (c),
    (a,b,c), (a,b), (a), ()
 )

```

````python
mc.select_data ('''
select 
     coalesce(os, '总计')            as os
    ,coalesce(device, '总计')        as device
    ,coalesce(city, '总计')          as city
    ,count(1)                        as cnt
from requests 
group by rollup (os, (os,device), city)
''')
````

通过group by、cube、grouping sets组合产生grouping sets
```sql
select os,device, city, count(*)
from requests 
group by os, cube(os,device), grouping sets(city);
--等效于如下语句。
select os,device, city, count(*)
from requests 
group by grouping sets((os,device,city),(os,city),(os,device,city));
```

## grouping  
grouping sets结果中使用NULL充当占位符，导致您会无法区分占位符NULL与数据中真正的NULL。因此，MaxCompute为您提供了grouping。grouping接受一个列名作为参数，如果结果对应行使用了参数列做聚合，**返回0，此时意味着NULL来自输入数据**。否则返回1，此时意味着NULL是grouping sets的占位符。

````python
mc.select_data('''
select
  if(grouping(os) == 0, os, '总计')          as os,
  if(grouping(device) == 0, device, '总计')  as device,
  if(grouping(city) == 0, city, '总计')      as city, 
  count(*) as count 
from requests 
group by grouping sets((os, device), (city), ());
''')
````

````python
def explot_df(df, primary_key, explode_col, pat):
    '''
    df:要炸裂的df,要求df只有两个字段。
    primary_key：主键
    explode_col：需要炸裂的字段
    pat:explode_col中的分隔符
    '''
    df = df.reset_index(drop=True)
    L = []
    for i in range(0, len(df)):
        h = df.loc[[i]]
        h_explode = h[explode_col].str.split(pat, expand=True).T
        h_explode.insert(0, 'suibian', h[primary_key].values[0])
        h_explode.columns = df.columns
        L.append(h_explode)
        
    return pd.concat(L)
````
